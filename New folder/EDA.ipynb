```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
# Define a function for easy plotting
# Plotting each variable against 'cnt'
def plot_relationship(x, y, data = day_Bike, hue = None, figsize = (10, 6), title = None, xlabel = None, 
                      ylabel = None):
    plt.figure(figsize = figsize)
    plot = sns.scatterplot(x = x, y = y, data = data, hue = hue)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
plot_relationship('temp', 'cnt', title='Temperature vs Bike Rentals', 
                  xlabel='Normalized Temperature', ylabel='Count of Rentals')
```

```python
plot_relationship('atemp', 'cnt', title='Feels-Like Temperature vs Bike Rentals', 
                  xlabel='Normalized "Feels-Like" Temperature', ylabel='Count of Rentals')
```

```python
plot_relationship('hum', 'cnt', title='Humidity vs Bike Rentals', 
                  xlabel='Normalized Humidity', ylabel='Count of Rentals')
```

```python
plot_relationship('windspeed', 'cnt', title='Wind Speed vs Bike Rentals', 
                  xlabel='Normalized Wind Speed', ylabel='Count of Rentals')
```

```python
plot_relationship('mnth', 'cnt', title='Month vs Bike Rentals', 
                  xlabel='Month', ylabel='Count of Rentals')
```

```python
day_Bike['dteday'] = pd.to_datetime(day_Bike['dteday'])
day_Bike['day'] = day_Bike['dteday'].dt.day
day_Bike['month'] = day_Bike['dteday'].dt.month
day_Bike['year'] = day_Bike['dteday'].dt.year
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(12, 6))
day_Bike['date'] = pd.to_datetime(day_Bike[['year', 'month', 'day']])
plt.scatter(day_Bike['date'], day_Bike['cnt'], color='blue', alpha=0.5)
plt.title('Bike Rentals Over Time')
plt.xlabel('Date')
plt.ylabel('Count (cnt)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns  
import pandas as pd

plt.figure(figsize=(10, 6))
# Plot casual against season using a bar plot
sns.barplot(x='season', y='casual', data=day_Bike, palette='viridis')
plt.title('Casual Bike Rentals by Season')
plt.xlabel('Season')
plt.ylabel('Count (casual)')
plt.show()
```
