```python
def kelvin_to_celsius_fahrenheit(kelvin):
    celsius = kelvin - 273.15
    fahrenheit = celsius * (9/5) + 32
    return celsius, fahrenheit
```

```python
temp_kelvin = response["main"]["temp"]
temp_celsius, temp_fahrenheit = kelvin_to_celsius_fahrenheit(temp_kelvin)
temp_celsius
```

```python
import requests
from datetime import datetime, timedelta

BASE_URL = "http://api.openweathermap.org/data/2.5/onecall/timemachine"
API_KEY = "c5610bb0ad50f6ddb52e2276182d31a2"
CITY = "Los Angeles"
LATITUDE = 34.0522
LONGITUDE = -118.2437

# Retrieve historical weather data for the past 5 days
for days_ago in range(10):
    date = (datetime.now() - timedelta(days=days_ago)).strftime("%Y-%m-%d")
    timestamp = int(datetime.strptime(date, "%Y-%m-%d").timestamp())

    api_url = f"{BASE_URL}?lat={LATITUDE}&lon={LONGITUDE}&dt={timestamp}&appid={API_KEY}"

    response = requests.get(api_url)
    weather_data = response.json()

    # Print the historical weather data for the current date
    print(f"Date: {date}")
    print("Weather Data:", weather_data)
    print("=" * 50)
```
