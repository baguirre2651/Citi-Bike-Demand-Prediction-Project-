```python
import pandas as pd
import os

# Directory where the combined_data.csv file is located
directory = value_target_directory

# File name
file_name = "combined_data.csv"

# Create the full path to the CSV file
csv_file_path = os.path.join(directory, file_name)

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(csv_file_path)

# Now, you can work with the DataFrame 'df' as needed
df.head(5)
```

```python
drop_columns = ['3000', 'Virtual Station', '7/7/2016', 'N/A', 'Active',
                'start station name', 'end station name', 'Unnamed: 0',
                'start_station_id', 'end_station_id']

# Drop the specified columns from the DataFrame
bike_sharing_data = df.drop(columns=drop_columns)
bike_sharing_data
```

```python
import pandas as pd

# Assuming your DataFrame is named bike_sharing_data
# Convert "start_time" to datetime format with error handling
bike_sharing_data['start_time'] = pd.to_datetime(bike_sharing_data['start_time'], errors='coerce')

# Create a new column with day, month, and year
bike_sharing_data['date_only'] = bike_sharing_data['start_time'].dt.strftime('%m/%d/%Y')

# Display the updated DataFrame
(bike_sharing_data.head())
```
